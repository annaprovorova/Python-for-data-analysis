В библиотеке pandas определены два класса объектов для работы с данными:

*Series* — одномерный массив, который может хранить значения любого типа данных;

*DataFrame* — двумерный массив (таблица), в котором столбцами являются объекты класса Series.

В pandas можно создать датафрейм и вручную:

```python
#s = pd.Series(data, index=index) 
#data могут выступать: массив numpy, словарь, число
#Подробнее о numpy еа следующей практике, сейчас используем обычный список

s = pd.Series([1,2,3,4,5], index=["a", "b", "c", "d", "e"])
print(s)
d = {"a": 10, "b": 20, "c": 30, "g": 40}
print(pd.Series(d))
print()
print(pd.Series(d, index=["a", "b", "c", "d"]))
index = ["a", "b", "c"]
print(pd.Series(5, index=index))
```

Для Series доступно взятие элемента по индексу, срезы, поэлементные математические операции аналогично массивам numpy.

```python
s = pd.Series([1, 2, 3, 4, 5], index=["a", "b", "c", "d", "e"])
print("Выбор одного элемента")
print(s["a"])
print("Выбор нескольких элементов")
print(s[["a", "d"]])
print("Срез")
print(s[1:])
print("Поэлементное сложение")
print(s + s)
```

В DataFrame есть несколько способов для обращения к строкам, столбцам и отдельным элементам таблицы: квадратные скобки и методы loc, iloc.

Как обычно, лучший источник знаний об этом — документация. Ниже краткое содержание.

В метод loc можно передать значение индекса (число, которое стоит в колонке index) строки, чтобы получить эту строку:

```python
data.loc[2]
```

```python
type(data.loc[2])
```

А с помощью срезов можно выбрать часть таблицы:

```python
data.loc[2:4]
```

Срез в loc производится по index и включает в себя последний элемент.

Метод iloc действует похожим образом, но он индексирует элементы не по index, а по порядку в таблице (который может отличаться от index). Например:

```python
subset = data.sample(5)
subset
```

```python
# subset.loc[2]
```

С помощью iloc тоже можно делать срезы, но в них последний элемент не включается (как и в обычных срезах в Python, в отличие от loc):

```python
subset.iloc[2:4]
```

Срезы можно брать не только по строкам, но и по столбцам. Обратите внимание на различия индексации столбцов в loc и iloc:

```python
data.iloc[2:4, 2:6]
```

```python
data.loc[2:4, 'Smokers.Total':'Smokers.Male']
```

Также для разведочного анализа может помочь метод groupby(column).

Он группирует объекты по указанной(-ым) колонке(-ам). Необходимо также указать какую статистику для группировки выводить. Это может быть количество (count), среднее (mean) или другие. Из огромной функциональности этого метода разберем только несколько базовых приемов:

```python
data.groupby('Year').sum()[['Smokers_Male', 'Smokers_Female']]
```

Попробуйте посмотреть средние значения по тем же группам и столбцам:

Что выводится в следующей строке?

```python
data.groupby(by='Country').count()[['Male', 'Female']].sum(axis=1)
```

Также данные иногда полезно отсортировать. Сейчас данные отсортированы в алфавитном порядке по странам. Отсортируем по возрастани. по общему количеству курильщиков:

```python
data['Total'] = data.Male + data.Female
data.sort_values(by="Total").head(3) #можно сохранить эти данные в новую переменную
```

А вот сортировка по убыванию:

```python
data.sort_values(by="Total", ascending=False)
```

Наконец, пожалуй, было бы интересно проиллюстрировать некоторые зависимости с помощью графиков. Для этого мы будем использовать библиотеку [matplotlib](https://matplotlib.org/).

Например, давайте попробуем посчитать ТОП-10 стран по количеству курильщиков за период 1980-2012.

```python
new_df = data.groupby(by='Country').sum().sort_values(by='Total', ascending=False).head(11)
new_df = new_df[1:]
new_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.bar(new_df.index, new_df['Total'])

plt.xticks(rotation=45)
plt.xlabel("Страна")
plt.ylabel("Количество курильщиков")
plt.title("ТОП-10 стран по количеству курильщиков"); # обратите внимание на точку с запятой
```

Постройте такую же диаграмму но только по 1989 году:

Теперь построим диаграмму, отображающую количество курящих людей в России:

```python
plt.figure(figsize=(10, 6))

plt.plot(
    data[data['Country']  == 'Russia'].Year,
    data[data['Country']  == 'Russia'].Total,
    
    color="blue"
)

plt.xlabel("Год")
plt.ylabel("Количество")
plt.title("Количество курящих людей в России с 1980-2012");
```

Попробуйте построить сами подобную диаграмму по дугой стране:

Давайте построим ещё одну диаграмму сравнение количества курильщиков-мужчин и женщин в России. Сделаем это с помощью круговой диаграммы.

```python
# точечная диаграмма

plt.pie([data[data.Country == 'Russia'].Male.sum(), data[data.Country == 'Russia'].Female.sum()],
        labels=['Male', 'Female'],
        autopct="%.1f%%",
        explode=[0, 0.1])

plt.title("Для курильщиков мужчин и женщин в России");
```
